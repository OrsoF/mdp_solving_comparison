```python
import pandas as pd

df = pd.read_csv('build.csv', index_col=False)
df = df.transpose()
mapper = {i : list(df.iloc[0])[i] for i in range(len(df.columns))}
df = df.rename(columns = mapper)
df = df.drop('Unnamed: 0')
df.plot(logy=True)
```
